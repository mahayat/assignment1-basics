```python
chunk = "the cat ate"
```

```python
import regex as re
from cs336_basics.bpe_training import PAT

words = []
for match in re.finditer(PAT, chunk):
    word = match.group()
    words.append(word)
    # word_bytes = tuple(bytes([b]) for b in word.encode('utf-8'))
```

```python
words
```

```python
vocab = {0: b' ', 1: b'a', 2: b'c', 3: b'e', 4: b'h', 5: b't', 6: b'th', 7: b' c', 8: b' a', 9: b'the', 10: b'at'}
inv_vocab = {v: k for k, v in vocab.items()}
```

```python
merges = [(b't', b'h'), (b' ', b'c'), (b' ', 'a'), (b'th', b'e'), (b' a', b't')]
merges
```

```python
word = words[0]
to_merge = [i.encode() for i in word]
to_merge
```

```python
n = len(to_merge)
i = 0

while len(to_merge) == n:
    while i < n:
        if i < n-1:
            to_merge_next = []
            if inv_vocab[(to_merge[i]+to_merge[i+1])]:
                to_merge_next.append(to_merge[i]+to_merge[i+1])
                i += 2
            else:
                to_merge_next.append(to_merge[i])
                i += 1
        else:
            to_merge_next.append(to_merge[i])
            i += 1
    to_merge = to_merge_next
```
